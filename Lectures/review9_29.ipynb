```r
library('tidyverse')
library('palmerpenguins')
```

```r
penguins |> head()
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm)) +
    geom_histogram(na.rm = TRUE, binwidth = 1) +
    geom_rug()
# na.rm gets rid of warning about missing values
p
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm, y = species)) +
    geom_point(na.rm = TRUE, position = position_jitter(height=0.2, width=0), shape=21)
# shape gives open circles, similar to transparency
# not specifying position jitter gives default values (not the same as setting it equal to zero)
p
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm, y = species)) +
    geom_boxplot(na.rm = TRUE, outlier.alpha = 0, alpha=0, color='grey40') + 
    geom_point(na.rm = TRUE, 
               shape = 21,
               color = 'steelblue',
               position = position_jitter(height = 0.2, width = 0))
#outlier.alpha=0 means making the point transparent to make them disappear
p
```

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = bill_length_mm, fill=species)) +
    geom_histogram(na.rm = TRUE, binwidth = 1, position='identity', alpha=0.5) +
    geom_rug(aes(color = species)) + scale_color_brewer(palette = 'Set2') + 
    scale_fill_brewer(palette = 'Set2')
# alpha allows u to see each group after doing position
p
```
