```r
library('tidyverse')
```

```r
df.marathon = read.csv('../data//marathon.csv')
df.marathon |> head()
```

Use `arrange` to find the best race time.

```r
df.marathon |>
arrange(race.time) |> head(1)
```

What are the best race times by country?

```r
df.marathon |>
group_by(country) |>
summarize(best.race.time = min(race.time)) 
```

In what year was the difference between the men's and women's winning time smallest? Use this table:

```r
df = df.marathon |>
     select(year, MF, race.time) |>
     tidyr::spread(MF, race.time) |>
    #filter(!is.na("Women's")) |> head() |> same as drop_na()
     drop_na() |>
     rename(men = `Men's`, women = `Women's`)

df |> head()
```

```r
df |>
mutate(time.difference = women - men) |>
arrange(time.difference) |> head(1)
#summarize(smallest.difference = min(time.difference)) 
```

Plot the difference between men's and women's race times by year.

```r
options(repr.plot.res = 250, repr.plot.width = 5, repr.plot.height = 4)
df |>
mutate(time.difference = abs(women - men)) |> #can also do abs value(men-women)
arrange(time.difference) |> head(1) #can make descending order by sorting by neg of col by adding - in front
#pipe operation %>% (older way)
```

```r
options(repr.plot.res = 250, repr.plot.width = 5, repr.plot.height = 4)
df_diff = df |> mutate(time.difference = abs(men-women))
p = ggplot(df_diff, aes(x=year, y = time.difference)) + geom_line() + geom_point()
p
```
