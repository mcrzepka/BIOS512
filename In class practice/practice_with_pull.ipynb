```r
library('tidyverse')
```

```r
p = ggplot(mpg, aes(y=manufacturer)) +
    geom_bar()

p
```

```r
mpg |>
    mutate(manufacturer = factor(manufacturer)) |>
    pull(manufacturer) #the levels sorts it from most frequent to least frequent
```

```r
mpg2 = mpg |>
    mutate(manufacturer = fct_infreq(manufacturer) |> fct_rev())

p = ggplot(mpg2, aes(y = manufacturer)) + geom_bar()

p
```
