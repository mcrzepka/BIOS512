Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width=4.5, repr.plot.height=5)
q = ggplot(penguins, aes(x=body_mass_g)) + 
    geom_histogram(fill='steelblue', alpha=0.75, color='steelblue', bins=37) + 
    facet_wrap(~year, ncol=1)+
    theme_minimal() + 
    scale_x_continuous(labels=comma_format()) +
    labs(x='Body mass in grams', y='Number of penguins', title='Histograms of penguin body mass for measurements taken from 2007-2009') +
    theme(axis.text = element_text(size = 6)) +
    theme(plot.title = element_text(size = 8)) +
    theme(axis.title = element_text(size = 8)) 

q
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
n = ggplot(penguins, aes(x=body_mass_g, y=sex)) + 
    geom_point(position = position_jitter(h = 0.25, w = 0.25), size=1, alpha=0.35) +
    facet_wrap(~species, ncol=1) +
    theme_minimal() + 
    labs(x='Body mass in grams', y='') + 
    scale_x_continuous(labels=comma_format())
n
```

Yes, there is a difference between the body mass of male and female penguins. The male penguins generally have a greater body mass than the female penguins across all species.
